Recipe 3: Data Visulization with Matplotlib

We will learn how to crate a simple plot to visuliaze data using Matplotlib

```python
import pandas as pd
import matplotlib. pyplot as plt
# This code will import the Matplotlib library's pyplot module and give it the alias plt.
#Matplotlib is a popular plotting library in Python
```

```python
df= pd.DataFrame({
    "year":[2015,2016,2017,2018,2019],
    "Sales":[200,250,270,350,500]
    #df = pd.DataFrame(...): Create a sample DataFrame with columns "year" and "Sales" containing some data.
})
#plotting
plt.plot(df["year"], df["Sales"])
#plt.plot(df["year"], df["Sales"]): Create a line plot using the "year" column as the x-axis and the "Sales" column as the y-axis.
plt.title("yearly Sales")
#plt.title("Yearly Sales"): Set the title of the plot to "Yearly Sales".
plt.xlabel("Year")
#plt.xlabel("Year"): Label the x-axis as "Year".
plt.ylabel("Sales")
#plt.ylabel("Sales"): Label the y-axis as "Sales".
plt.show()
#plt.show(): Display the plot.
```
